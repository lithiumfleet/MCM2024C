```python
from pandas import read_csv
import matplotlib.pyplot as plt
```

```python
df = read_csv('./2024c/all_data_in_int.csv', index_col=[0])
```

```python
p1_colums, p2_colums = [], []
others = []
for i in df.columns.tolist():
    i = i.strip()
    if i.find("p1") != -1 and i != 'p1_distance_run': p1_colums.append(i)
    elif i.find("p2") != -1 and i != 'p2_distance_run': p2_colums.append(i)
    else: others.append(i)
print(p1_colums)
```

```python
dflist = [df[p1_colums], df[p2_colums], df[others]]
for i in dflist:
    i = (i-i.mean()) / i.std()
    # i.plot()
```
