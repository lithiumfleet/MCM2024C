```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('./2024c/data/preprocessmatches.csv')
df.isnull().sum()
# speed_mph             11
# serve_width            1
# serve_depth            1
# return_depth          48
```

```python
# 删除后面四个列便于分析
need_to_drop = ['winner_shot_type','match_id','elapsed_time','serve_width','serve_depth','speed_mph','serve_width','serve_depth','return_depth']
df = df.drop(labels=need_to_drop, axis=1)
# AD无法量化, 现在把AD换成50
df = df.replace('AD', 50)
df.to_csv('2024c/all_data_in_int.csv')
df.head()
```

```python
sns.heatmap(df.corr())
plt.figure(figsize=(35,30))
plt.xticks(fontsize=50)
plt.yticks(fontsize=50)
plt.show()
```
